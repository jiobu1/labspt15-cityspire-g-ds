```python
import pandas as pd
 
train = pd.read_csv(r'https://raw.githubusercontent.com/Med-Cabinet-Project/DS/master/web_app/stats_model/strains.csv', header='infer', sep=',')
train.head()
```

```python
train.describe(exclude='number')
```

```python
x = train.medical.str.split('\s*,\s*', expand=True).stack()
pd.crosstab(x.index.get_level_values(0), x.values).iloc[:, 1:]
```

```python
x = train.positive.str.split('\s*,\s*', expand=True).stack()
pd.crosstab(x.index.get_level_values(0), x.values).iloc[:, 1:]
```
